```python
# Access the elevation DEM (Digital Elevation Model) using arcpy
import arcpy

elevationLayer = "C:/EsriTraining/ProStart/NorthridgeData/Database/Northridge.gdb/Elevation"

arcpy.Exists(elevationLayer)

# Check out the ArcGIS Spatial Analyst extension license
arcpy.CheckOutExtension("Spatial")
```

```python
# Conduct slope analysis
outSlope = arcpy.sa.Slope(elevationLayer, "DEGREE")

# Save the output
outSlope.save("C:/EsriTraining/myFCs.gdb/outSlope")
```

```python
# Conduct aspect analysis
outAspect = arcpy.sa.Aspect(elevationLayer)

# Save the output
outAspect.save("C:/EsriTraining/myFCs.gdb/outAspect")
```

```python
# Sample arithmetic operations
arithmeticLayer = outSlope * 10 + (outAspect - 15 ) / 2

arithmeticLayer.save("C:/EsriTraining/myFCs.gdb/arithmeric")
```

```python
# Sample arithmetic operations
arithmeticLayer2 = outSlope * 5 + (outAspect + 15 )

arithmeticLayer2.save("C:/EsriTraining/myFCs.gdb/arithmeric2")
```

```python
# Get the Slope layer from firestation.gdb\
# Look into its parameters

import arcpy

mySlope = "C:/EsriTraining/firestation.gdb/Slope"
arcpy.Exists(mySlope)

myRaster = arcpy.Raster(mySlope)

```

```python
print(myRaster.maximum)
```

```python
print("Minimum value: " + str(myRaster.minimum) + 
     "\nMaximum value: " + str(myRaster.maximum) +
     "\nWidth (pixels): " + str(myRaster.width) + 
     "\nHeight (pixels): " + str(myRaster.height) + 
     "\nMean cell width: " + str(myRaster.meanCellWidth) + 
     " in " + myRaster.spatialReference.linearUnitName)
```

```python
print("The number of pixels in this layer: " + 
      str(myRaster.width * myRaster.height * 
          myRaster.meanCellWidth * myRaster.meanCellWidth /
         1.076e+7))
```

```python
# Convert raster to numpy array
arr = arcpy.RasterToNumPyArray(myRaster, nodata_to_value = 0)
```

```python
print(arr)
```

```python
import matplotlib.pyplot as plt
imgplot = plt.imshow(arr, cmap='gray')
```
